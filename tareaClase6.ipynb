Calculadora de IMC (Índice de Masa Corporal):

    Cree una función que calcule el IMC (Índice de Masa Corporal) de una persona a partir de su peso y altura. La función debe recibir como parámetros el peso (en kg) y la altura (en cm) y devolver el valor del IMC.

Opcional: Implemente una segunda función que clasifique el IMC según los estándares de la OMS:

    Bajo peso: IMC < 18.5
    Peso normal: 18.5 <= IMC < 25
    Sobrepeso: 25 <= IMC < 30
    Obesidad: IMC >= 30

```python
#Funcion que calcula el IMC (IMC = peso (kg)/ [estatura (m)]2)
def calcularIMC(peso,altura_cm):
    altura_m = altura_cm / 100 #Pasamos la altura de cm a metros
    imc = peso/(altura_m**2) #calculo para sacar el imc
    return round(imc, 1) #Devuelvo el valor imc redondeado para que solo muestre un decimal

#Funcion que clasifica el IMC segun la OMS
def clasificacionIMC(imc):
    if(imc < 18.5):
        clasificacionOMS = "Bajo peso"
    elif(18.5 <= imc < 25):
        clasificacionOMS = "Peso normal "
    elif(25 <= imc < 30):
        clasificacionOMS = "Sobrepeso"
    else:
        clasificacionOMS = "Obesidad"
    return clasificacionOMS

# -------------- Codigo Principal  --------------
#Pido al usuario los datos necesarios
pesoUsuario = float(input("Ingrese su peso(Kg): "))
alturaUsuario = float(input("Ingrese su altura(cm): "))

#Guardo en variables los valores calculados en las funciones
imc = calcularIMC(pesoUsuario,alturaUsuario)
clasificacion =  clasificacionIMC(imc)

#Imprimo por pantalla los resultados
print("Tu IMC es: ", imc, ", Tu clasificacion es: ", clasificacion)

```

Analizador de números:

    Cree una función que reciba un número entero como parámetro y devuelva un diccionario con la siguiente información:

        Si el número es par o impar
        Si el número es positivo, negativo o cero
        Si el número es divisible por 3
        Si el número es divisible por 5
        Si el número es divisible por 3 y por 5
Opcional: Implemente la función utilizando una función lambda para cada una de las verificaciones.

```python
#Funcion diccionario, calcula la informacion del numero del usuario
def diccionarioNum (numero):
    #Funciones lambda para cada una de las verificaciones
    par_impar = lambda num:  "par" if(num % 2 == 0 ) else "impar" 
    divisible_por_3 = lambda num: "si" if (num % 3 == 0) else "no"
    divisible_por_5 = lambda num: "si" if (num % 5 == 0) else "no"
    divisible_por_3y5 = lambda num: "si" if (divisible_por_3(num)=="si" and divisible_por_5(num)=="si") else "no"

    #Positivo negativo o cero
    if(numero > 0):
        valorNumero = "Positivo"
    elif(numero < 0):
        valorNumero = "Negativo"
    else:
        valorNumero = "es cero"

    #Creo el diccionario
    info_Num = {
        "El numero es": par_impar(numero),
        "El signo del numero es": valorNumero,
        "DivisiblePor_3": divisible_por_3(numero),
        "DivisiblePor_5": divisible_por_5(numero),
        "DivisiblePor_3y5": divisible_por_3y5(numero)

    }

    return(info_Num) #Devuelvo el diccionario

# ---------- Codigo Principal ----------

#Pido al usuario un numero entero
numUsuario = int(input("Ingrese un numero entero: "))
#muestro por pantalla el diccionario 
print(diccionarioNum(numUsuario))
```

Calculadora de área y perímetro de figuras geométricas:

    Cree funciones para calcular el área y perímetro de las siguientes figuras geométricas:

        Cuadrado [area = lado al cuadrado | perimetro = lado*4]
        Triángulo (equilátero[area = (base * altura)/ 2 | perimetro = lado* 3], rectángulo[ area = ] e isósceles)
        Círculo [area = pi * r al cuadrado | perimetro = 2 * pi * r]
        Rectángulo [area = base * altura  | perimetro = 2*(base + lado)]
Cada función debe recibir las dimensiones necesarias como parámetros y devolver el valor correspondiente del área o perímetro.

```python
#Devuelve el area y perimetro de un cuadrado
def cuadrado(lado):
    area = lado * lado
    perimetro = lado * 4
    return area, perimetro

#Devuelve el area y perimetro de un rectangulo
def rectangulo(largo, ancho):
    area = largo * ancho
    perimetro = 2 * (largo + ancho)
    return area, perimetro

#Devuelve el area y perimetro de un circulo
def circulo(radio):
    area = 3.1416 * (radio ** 2)
    perimetro = 2 * 3.1416 * radio
    return area, perimetro

#Devuelve el area y perimetro de un triangulo equilatero
def triangulo_equilatero(lado):
    area =  (lado**2 * (3**0.5)) / 4
    perimetro = lado * 3
    return area, perimetro

#Devuelve el area y perimetro de un triangulo rectangulo
def triangulo_rectangulo(base, altura, lado1, lado2):
    area = (base * altura) / 2
    perimetro = lado1 + lado2 + base
    return area, perimetro

#Devuelve el area y perimetro de un triangulo isosceles
def triangulo_isosceles(base, altura, lados):
    area = (base * altura) / 2
    perimetro = (lados*2) + base
    return area, perimetro

#Pide los valores necesarios para el calculo y muestra por pantalla el area y el perimetro 
def opcionesMenu(opcion):
    if opcion == 1:# Cuadrado
        lado = float(input("Ingrese el lado del cuadrado: "))
        area, perimetro = cuadrado(lado)
        print("El area del cuadrado es: ", area, "y su perimetro es: ", perimetro)

    elif (opcion == 2):# Rectangulo
        largo = float(input("Ingrese el largo del rectangulo: "))
        ancho = float(input("Ingrese el ancho del rectangulo: "))
        area, perimetro = rectangulo(largo, ancho)
        print("El area del rectangulo es: ", area, "y su perimetro es: ", perimetro)

    elif (opcion == 3):# Circulo
        radio = float(input("Ingrese el radio del circulo: "))
        area, perimetro = circulo(radio)
        print("El area del circulo es: ", area, "y su perimetro es: ", perimetro)

    elif (opcion == 4):#Triangulo Equilatero
        lado = float(input("Ingrese el lado del triangulo equilatero: "))
        area, perimetro = triangulo_equilatero(lado)
        print("El area del triangulo equilatero es: ", area, "y su perimetro es: ", perimetro)

    elif (opcion == 5):#Triangulo Rectangulo
        lado1 = float(input("Ingrese un lado del triangulo rectangulo: "))
        lado2 = float(input("Ingrese otro lado del triangulo rectangulo: "))
        base = float(input("Ingrese la base del triangulo rectangulo: "))
        altura = float(input("Ingrese la altura del triangulo rectangulo: "))
    
        area, perimetro = triangulo_rectangulo(base, altura, lado1, lado2)
        print("El area del triangulo rectangulo es: ", area, "y su perimetro es: ", perimetro)

    elif (opcion == 6):#Triangulo Isosceles
        lados = float(input("Ingrese un lado del triangulo isosceles: "))
        base = float(input("Ingrese la base del triangulo isosceles: "))
        altura = float(input("Ingrese la altura del triangulo isosceles: "))
        area, perimetro = triangulo_isosceles(base, altura, lados)
        print("El area del triangulo isosceles es: ", area, "y su perimetro es: ", perimetro)

    else:
        print("Opcion no valida")

#Muestra por pantalla las diferentes figuras que se puede calcular el area y perimetro 
def menu():
    print("De que figura quiere el area y el perimetro?")
    print(" 1. Cuadrado")
    print(" 2. Rectangulo")
    print(" 3. Circulo")
    print(" 4. Triangulo equilatero")
    print(" 5. Triangulo rectangulo")
    print(" 6. Triangulo isosceles")
    opcion = int(input("Ingrese la opcion: "))
    print("----------------------------------------------")
    opcionesMenu(opcion)
        
#------- Codigo principal -------

#Llamo a la funcion menu
menu()





```
